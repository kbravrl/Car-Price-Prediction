```python
df = df[df.year != 1970]
```

```python
model = pd.get_dummies(df['model'], dtype='int')
model.columns = [model + "_model" for model in model.columns]

transmission = pd.get_dummies(df['transmission'], dtype='int')
transmission.columns = [transmission + "_transmission" for transmission in transmission.columns]

fuelType = pd.get_dummies(df['fuelType'], dtype='int')
fuelType.columns = [fuelType + "_fuelType" for fuelType in fuelType.columns]

df = pd.concat([df, model, transmission, fuelType], axis=1)

df = df.drop(['model', 'transmission', 'fuelType'], axis=1)
```

```python
df.corr()['price']
```

```python
X= df.drop('price', axis=True).values
y= df['price'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train_s = scaler.fit_transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(40, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(40, activation='relu'))

model.add(Dense(1))

model.compile(optimizer= 'adam', loss='mse')
```

```python
model.fit(x=X_train_s, y=y_train, validation_data=(X_test_s, y_test), batch_size= 250, epochs=300)
```

```python
loss_data = pd.DataFrame(model.history.history)
```

```python
loss_data.head()
```

```python
loss_data.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
y_predict = model.predict(X_test_s)
```

```python
y_predict
```

```python
mean_absolute_error(y_test, y_predict)
```
